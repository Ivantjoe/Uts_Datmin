```python
# prompt: buatkan EDA (univariate dan multivariate
# analysis, minimal 3 analisis tiap pendekatan)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    data = pd.read_excel('/content/Data BPS.xlsx')

    # Data Cleaning (you might need to adjust this based on your data)
    for col in data.columns:
        if pd.api.types.is_numeric_dtype(data[col]):
            data[col] = data[col].fillna(data[col].mean())
        else:
            data[col] = data[col].fillna(data[col].mode()[0])
    data.drop_duplicates(inplace=True)

    # Univariate Analysis
    # 1. Descriptive Statistics
    print("Descriptive Statistics:\n", data.describe())

    # 2. Histograms
    for col in data.select_dtypes(include=['number']).columns:  # Iterate numeric columns
        plt.figure()  # Create a new figure for each histogram
        sns.histplot(data[col], kde=True) # KDE adds a kernel density estimate
        plt.title(f'Distribution of {col}')
        plt.show()

    # 3. Value Counts for Categorical Variables
    for col in data.select_dtypes(exclude=['number']).columns: # Iterate categorical columns
        print(f"Value counts for {col}:\n{data[col].value_counts()}\n")
        plt.figure() # New figure for each barplot
        data[col].value_counts().plot(kind='bar')
        plt.title(f'Value Counts of {col}')
        plt.show()


    # Multivariate Analysis
    # 1. Correlation Matrix
    numeric_data = data.select_dtypes(include=['number'])
    correlation_matrix = numeric_data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

    # 2. Scatter Plots (example with two numerical variables)
    # Replace 'variable1' and 'variable2' with your column names
    if 'variable1' in data.columns and 'variable2' in data.columns:
        plt.figure()
        sns.scatterplot(x='variable1', y='variable2', data=data)
        plt.title('Scatter Plot of variable1 vs variable2')
        plt.show()

    # 3. Boxplots (example with a numerical variable grouped by a categorical variable)
    # Replace 'numerical_var' and 'categorical_var' with your column names
    if 'numerical_var' in data.columns and 'categorical_var' in data.columns:
        plt.figure()
        sns.boxplot(x='categorical_var', y='numerical_var', data=data)
        plt.title('Boxplot of numerical_var by categorical_var')
        plt.show()

except FileNotFoundError:
    print("Error: 'Data BPS.xlsx' not found. Please provide a valid file path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# prompt: buatkan clustering dengan KMeans

from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

try:
    data = pd.read_excel('/content/Data BPS.xlsx')

    # Data Cleaning (you might need to adjust this based on your data)
    for col in data.columns:
        if pd.api.types.is_numeric_dtype(data[col]):
            data[col] = data[col].fillna(data[col].mean())
        else:
            data[col] = data[col].fillna(data[col].mode()[0])
    data.drop_duplicates(inplace=True)

    # Select numerical features for clustering
    numeric_data = data.select_dtypes(include=['number'])

    # Determine optimal number of clusters (e.g., using the Elbow method)
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(numeric_data)
        wcss.append(kmeans.inertia_)

    plt.plot(range(1, 11), wcss)
    plt.title('Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()

    # Apply KMeans clustering (replace 'optimal_k' with the value you determined)
    optimal_k = 3 # Example, change this based on the elbow method plot
    kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
    data['cluster'] = kmeans.fit_predict(numeric_data)

    # Print the clustered data
    print(data)

    # Visualization (example with two features)
    if len(numeric_data.columns) >=2:
      plt.scatter(data['variable1'], data['variable2'], c=data['cluster'], cmap='viridis')
      plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
      plt.title('Clusters')
      plt.xlabel('variable1') #replace with your column name
      plt.ylabel('variable2') #replace with your column name
      plt.legend()
      plt.show()

except FileNotFoundError:
    print("Error: 'Data BPS.xlsx' not found. Please provide a valid file path.")
except KeyError:
    print("Error: 'variable1' or 'variable2' column not found. Check your column names.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# prompt: buatkan pca

from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

try:
    data = pd.read_excel('/content/Data BPS.xlsx')

    # Data Cleaning (you might need to adjust this based on your data)
    for col in data.columns:
        if pd.api.types.is_numeric_dtype(data[col]):
            data[col] = data[col].fillna(data[col].mean())
        else:
            data[col] = data[col].fillna(data[col].mode()[0])
    data.drop_duplicates(inplace=True)

    # Select numerical features for PCA
    numeric_data = data.select_dtypes(include=['number'])

    # Apply PCA
    pca = PCA(n_components=2)  # Reduce to 2 principal components
    principal_components = pca.fit_transform(numeric_data)

    # Create a new DataFrame with the principal components
    pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

    # Explained variance ratio
    explained_variance_ratio = pca.explained_variance_ratio_
    print(f"Explained Variance Ratio:\n{explained_variance_ratio}")

    # Visualize the principal components
    plt.figure(figsize=(8, 6))
    plt.scatter(pca_df['PC1'], pca_df['PC2'])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('PCA of Numerical Data')
    plt.show()

except FileNotFoundError:
    print("Error: 'Data BPS.xlsx' not found. Please provide a valid file path.")
except Exception as e:
    print(f"An error occurred: {e}")
```
